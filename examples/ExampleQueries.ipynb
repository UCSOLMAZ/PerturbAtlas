# PerturbAtlas Example Queries

This notebook demonstrates several practical ways to use **PerturbAtlas-K562-v1.0**.

It includes:

1. Loading the atlas.
2. Querying known perturbation-derived gene subsets.
3. Comparing myeloid, erythroid, and developmental signatures.
4. Testing lowercase and duplicated gene inputs.
5. Testing low-evidence/random-like queries.
6. Exporting top perturbation results.

No external example data files are required. Example gene sets are generated directly from the atlas.


## 1. Import and load the atlas

```python
from perturbatlas import K562Atlas

atlas = K562Atlas.load("../data/PerturbAtlas-K562-v1.0.pkl")

atlas.info()
```

## 2. Check available modules and perturbations

```python
print("Modules:")
print(atlas.modules())

print("\nFirst 10 perturbations:")
print(atlas.perturbations()[:10])
```

## 3. Example 1 — Myeloid-like query

Here we use the first 40 genes from the `CEBPA` perturbation signature.


```python
myeloid_query = list(atlas.strong_up["CEBPA"])[:40]

myeloid_result = atlas.query(myeloid_query)

myeloid_result.summary()
myeloid_result.interpret()
```

## 4. Example 2 — Erythroid-like query

Here we use the first 40 genes from the `KLF1` perturbation signature.


```python
erythroid_query = list(atlas.strong_up["KLF1"])[:40]

erythroid_result = atlas.query(erythroid_query)

erythroid_result.summary()
erythroid_result.interpret()
```

## 5. Example 3 — Developmental-like query

Here we use the first 40 genes from the `HOXA13` perturbation signature.


```python
developmental_query = list(atlas.strong_up["HOXA13"])[:40]

developmental_result = atlas.query(developmental_query)

developmental_result.summary()
developmental_result.interpret()
```

## 6. Compare evidence scores across examples

```python
scores = {
    "myeloid_CEBPA": myeloid_result.evidence_score(),
    "erythroid_KLF1": erythroid_result.evidence_score(),
    "developmental_HOXA13": developmental_result.evidence_score()
}

scores
```

## 7. Case normalization and duplicate handling

PerturbAtlas automatically normalizes gene symbols to uppercase and removes duplicate genes.


```python
messy_query = [
    "cebpa",
    "CEBPA",
    " spi1 ",
    "SPI1",
    "cebpe",
    "CEBPE"
]

messy_result = atlas.query(messy_query)

print("Cleaned query genes:")
print(messy_result.query_genes)

messy_result.summary()
```

## 8. Low-evidence example

A very small or weak query may not provide enough evidence for confident interpretation.


```python
weak_query = ["BAK1", "CEBPA", "SPI1", "KLF1"]

weak_result = atlas.query(weak_query)

weak_result.summary()
weak_result.interpret()
```

## 9. Export retrieved perturbations

The retrieved perturbation table can be exported for downstream analysis.


```python
myeloid_result.to_csv("example_myeloid_results.csv")

print("Saved: example_myeloid_results.csv")
```

## 10. Inspect top perturbations as a DataFrame

```python
myeloid_result.top_perturbations(n=10)
```

## Notes

- Use at least **20 informative genes** when possible.
- Low evidence scores should be interpreted conservatively.
- PerturbAtlas does not require black-box model predictions; every result is linked to retrieved perturbation signatures.
