# PerturbAtlas QuickStart

This notebook demonstrates the basic usage of **PerturbAtlas-K562-v1.0**.

You will learn how to:

1. Load the K562 perturbation atlas.
2. Inspect atlas information.
3. Query a gene set.
4. Summarize and interpret the result.
5. Export retrieved perturbations.
6. Explore available modules and perturbations.


## 1. Import PerturbAtlas

````python
from perturbatlas import K562Atlas
````

## 2. Load the atlas

The atlas file is expected to be stored in:

```text
data/PerturbAtlas-K562-v1.0.pkl
```


````python
atlas = K562Atlas.load("../data/PerturbAtlas-K562-v1.0.pkl")
````

## 3. Inspect atlas information

````python
atlas.info()
````

## 4. Define a query gene set

This example uses a small myeloid-associated gene set.


````python
query_genes = [
    "SPI1",
    "CEBPA",
    "CEBPE",
    "CSF3R",
    "MPO",
    "ELANE"
]
````

## 5. Query the atlas

````python
result = atlas.query(query_genes)
````

## 6. Summarize the result

````python
result.summary()
````

## 7. Interpret the biological program

````python
result.interpret()
````

## 8. View top perturbations as a table

````python
result.top_perturbations(n=10)
````

## 9. Export results

This saves the module-consistent perturbation retrieval table.


````python
result.to_csv("quickstart_results.csv")

print("Saved: quickstart_results.csv")
````

## 10. Explore available modules and perturbations

````python
atlas.modules()
````

````python
atlas.perturbations()[:10]
````

## Notes

- For best performance, use at least **20 informative genes** when possible.
- Low evidence scores should be interpreted conservatively.
- PerturbAtlas is intended as an interpretable perturbation-retrieval framework, not a black-box classifier.
